```python
deliveries = pd.read_csv('deliveries_updated_mens_ipl.csv')
```

```python
deliveries.head(2)
```

```python
def phase(over):
    if over <= 5:
        return 'Powerplay'
    elif over <= 14:
        return 'Middle'
    else:
        return 'Death'
```

```python
deliveries['phase'] = deliveries['over'].apply(lambda x: phase(x))
```
